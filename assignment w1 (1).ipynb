# 5.visualisation

# histogram , distribution of sepal width

```python


plt.figure(figsize=(8,8))
plt.hist(iris['SW'],color='blue',rwidth=0.9)
plt.title('Distribution of SW',fontsize=16)
plt.xlabel('Sepal width',fontsize=16)
plt.ylabel('Frequency',fontsize=16)
plt.xticks()
plt.yticks()
```

# histogram , distribution of sepal length

```python

plt.figure(figsize=(8,8))
plt.hist(iris['SL'],color='violet',rwidth=0.9)
plt.title('Distribution of SL',fontsize=16)
plt.xlabel('Sepal Length',fontsize=14)
plt.ylabel('Frequency',fontsize=14)
plt.xticks()
plt.yticks()
```

# distribution of petal length 

```python


plt.figure(figsize=(8,8))
plt.hist(iris['PL'],color='red',rwidth=0.9)
plt.title('Distribution of PL',fontsize=16)
plt.xlabel('Petal Length',fontsize=14)
plt.ylabel('Frequency',fontsize=14)
plt.xticks()
plt.yticks()
```

# distribution of petal width

```python


plt.figure(figsize=(8,8))
plt.hist(iris['PW'],color='green',rwidth=0.9)
plt.title('Distribution of PW',fontsize=16)
plt.xlabel('Petal Width',fontsize=14)
plt.ylabel('Frequency',fontsize=14)
plt.xticks()
plt.yticks()
```

# scatter plot of sepal width vs sepal length

```python


plt.figure(figsize=(8,6))
plt.subplot(2,1,1)
plt.scatter(iris['SL'],iris['SW'],s=10,marker='*',c='blue')
plt.title('Scatter plot of Sepal Lenth Vs Sepal Width',fontsize=16)
plt.xlabel('Sepal Length',fontsize=14)
plt.ylabel('Sepal Width',fontsize=14)


plt.subplot(2,1,2)
plt.scatter(iris['PL'],iris['PW'],s=10,marker='^',c='green')
plt.title('Scatter plot of Petal Length Vs Petal Width',fontsize=16)
plt.xlabel('Petal Length',fontsize=14)
plt.ylabel('Petal Width',fontsize=14)
plt.tight_layout()
```
